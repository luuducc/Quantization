```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# Chương trình lượng tử hóa tín hiệu sin xn
print("Chương trình lượng tử hóa xn dạng sin")
f = float(input("Nhập f0: "))
N = int(input("Nhập N: "))
A = int(input("Nhập biên độ A: "))
b = int(input("Nhập số bit lượng tử: "))
# A = 1
# f = 0.02
# N = 200

def Quantization(A,f,b,N):
    # Vẽ đồ thị xn
    n = np.arange(0,N-1,1)
    xt = A*np.sin(2 * math.pi * f * n)
    plt.stem(n,xt)
    plt.xlabel("Trục n")
    plt.ylabel("Trục biên độ")
    plt.title("Đồ thị xn")
    plt.axis([0,50, -1.1, 1.1]) 
    plt.show()
    # Hàm lượng tử hóa rounding
    def rounding(value,delta):
        result = float(value) / delta # tính giá trị của x/denta = result
        value = round(result) * delta # làm tròn result về số nguyên gần nhất, nhân lại với denta được giá trị x đã lượng tử
        return value # trả lại giá trị x đã được lượng tử
    # Vẽ đồ thị xqn
    xqn = []
    delta = 2*A/2**b
    for x in xt:
        xqn.append(rounding(x,delta)) # thêm giá trị đã được lượng tử hóa vào mảng xqn
    plt.stem(n,xqn,"r")
    plt.xlabel("Trục n")
    plt.ylabel("Trục biên độ")
    plt.title("Đồ thị xqn")
    plt.axis([0,50, -1.1, 1.1])
    plt.show()
    
    # Vẽ đồ thị eqn
    eqn = (xqn - xt)
    plt.stem(n,eqn,"y")
    plt.xlabel("Trục n")
    plt.ylabel("Trục biên độ")
    plt.title("Đồ thị eqn")
    plt.axis([0,50, -delta/2+0.001, delta/2+0.001])
    plt.show()    
    
    # Tính SQNR
    Px = A**2/2 
    eqn2 = eqn ** 2
    Pq = 0
    for x in eqn2:
        Pq += x
    Pq = Pq/N
    SQNR = Px/Pq
    SQNRlythuyet = 1.76 + 6.02 * b
    print("SQNR = ",math.log(SQNR)/math.log(10) * 10)
    print("SQNRlythuyet = ",SQNRlythuyet)
Quantization(A,f,b,N)
```
